```python
# packages
import numpy as np
import matplotlib.pyplot as plt
#Simpson's rule approximates the integral f(x) dx by the sum:
    #(h/3) \sum_{k=1}^{N/2} (f(x_{2i-2} + 4f(x_{2i-1}) + f(x_{2i}))
    #where x_i = a + i*dx and h = (b - a)/N.
def simpson(f, a, b, N = 50):
      if N % 2 == 1:
        raise ValueError("N must be an even integer.")
      h = (b-a)/N
      x = np.linspace(a, b, N+1)
      y = f(x)
      Sim = h/3 * np.sum(y[0:-1:2] + 4*y[1::2] + y[2::2])
      return Sim
```

Output of Numerical Integration

```python
simpson(np.sin,0,np.pi/2,100)
```
